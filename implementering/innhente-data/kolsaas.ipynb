```python
import matplotlib.pyplot as plt
```

```python
fil = open("kolsaas.csv")
data = fil.read()
```

```python
data
```

```python
linjer = data.split("\n")
tur = []
for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)
tur[1][0] # gir meg den første tiden, altså 0
```

```python
## Oppgave: plott time (tid) på x-aksen mt altitude (høyde) på y-aksen

x = []
y = []
lengde = []
bredde = []

for linje in tur[1:]: # betyr start på nummer 1, -1 betyr start på starten men ikke ta med siste
    høyde = float(linje[3])
    lengde.append(float(linje[2]))
    bredde.append(float(linje[1]))
    tid = float(linje[0])
    x.append(tid)
    y.append(høyde)

plt.scatter(x,y)
plt.show()
```
