```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = []
y1 = []

def f(x):
    return  3 * x + 2

for i in range(6):
    x1.append(i)
    y1.append(f(i))

plt.plot(x1,y1)
plt.scatter(x1, y1)
plt.show()
```
